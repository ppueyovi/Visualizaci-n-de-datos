Como podemos ver en el mapa superior, faltan países. Algunos de estos países los tengo en el dataset pero con nombre diferente, por lo que hago una conversión. Otros, no están en GeoPandas, o bien son países que en GeoPandas son cuidades capital. Por ejemplo -> Hong Kong

```python
deleteCountry = []
modifyCountry = []
paisCiudad = []

for pais in paises:
    if pais == 'United States':
        modifyCountry.append(pais)
    elif pais == 'Equatorial Guinea':
        modifyCountry.append(pais)
    elif pais == 'Dominican Republic':
        modifyCountry.append(pais)
    elif pais == 'South Sudan':
        modifyCountry.append(pais)
    elif pais == 'Bosnia And Herzegovina':
        modifyCountry.append(pais)
    elif pais == 'Solomon Islands':
        modifyCountry.append(pais)
    elif pais == 'Trinidad And Tobago':
        modifyCountry.append(pais)
    else:
        if not pais in paises_gpd:
            #COMPRUEBO SI ES UNA CIUDAD CAPITAL
            if pais in cities['name'].values:
                paisCiudad.append(pais)
            else:
                #GUARDO UNA LISTA CON EL PAIS QUE TENGO EN MI DATASET PERO QUE NO ES NI PAIS NI CAPITAL.
                deleteCountry.append(pais)

# PROCEDO A BORRAR DE MI DATASET AQUELLOS REGISTROS CUYO PAIS NO TENGO EN GEOPANDAS NI COMO PAIS NI COMO CAPITAL DE PAIS
df = df[~df.country.isin(deleteCountry)]
for pais in deleteCountry:
    df = df.drop(df[df['country']==pais].index)
# MODIFICO EL NOMBRE DE AQUELLOS PAISES QUE EN GEOPANDAS TIENEN NOMBRE -> United States -> United States of America
df.loc[ df['country'] == 'United States', 'country'] = 'United States of America'
df.loc[ df['country'] == 'Equatorial Guinea', 'country'] = 'Eq. Guinea'
df.loc[ df['country'] == 'Dominican Republic', 'country'] = 'Dominican Rep.'
df.loc[ df['country'] == 'South Sudan', 'country'] = 'S.Sudan'
df.loc[ df['country'] == 'Bosnia And Herzegovina', 'country'] = 'Bosnia and Herz.'
df.loc[ df['country'] == 'Solomon Islands', 'country'] = 'Solomon Is.'
df.loc[ df['country'] == 'Trinidad And Tobago', 'country'] = 'Trinidad and Tobago'

paises = df.country.unique()

world = world[world.name.isin(paises)]
m = world.explore(column='pop_est',cmap='Set2', legend=False, tooltip=['name'], zoom_control=False, dragging=True, scrollWheelZoom=True)

# BORRO DE CITIES AQUELLAS CIUDADES QUE NO ESTAN EN MI DATASET
cities = cities[cities.name.isin(paisCiudad)]
#PINTO EL MAPA CON LOS PAISES QUE VOY A ANALIZAR
mapa= cities.explore(
     m=m, # pass the map object
     color="red",
     marker_kwds=dict(radius=5, fill=True)
)


mapa
```

Voy a mostrtar una gráfica interactiva con el salario medio de cada país. Están en orden alfabético, lo que ayuda a la búsqueda del país que se quiere consultar.


```python
import matplotlib.pyplot as plt

import pandas as pd
import seaborn as sns

import ipywidgets as widgets
from ipywidgets import interact, interact_manual



countrySalary = df.groupby(['country'],as_index=False)[['Average Monthly Net Salary (After Tax) (USD)']].mean().round(2)
citySalary = pd.DataFrame()

def draw_barplot(a,pais,indice):
    #print(pais)
    #BUSCO EL SALARIO MEDIO DEL PAIS EN EL CORRESPONDIENTE DATAFRAME
    if indice:
        salario = countrySalary.iloc[indice, countrySalary.columns.get_loc('Average Monthly Net Salary (After Tax) (USD)')]
        #print(salario)
        info = 'EL SALARIO MEDIO MENSUAL EN ' + pais + ' SON ' + str(salario) + ' DOLARES'
        plt.title(info)


    cols = ['yellow' if (x == pais) else 'blue' for x in countrySalary.country ]
    sns.barplot(ax=a, x="country", y="Average Monthly Net Salary (After Tax) (USD)", data=countrySalary, picker=2, palette=cols)
    

def onpick(event):

    ax.cla()
    draw_barplot(ax,'','')
    ax.figure.canvas.draw_idle()

    idx = int(event.xdata.round())
    lbls = ax.get_xticklabels()
    lbl = lbls[idx]
    salario = ax.get_yticklabels()
    pais = lbl.get_text()

    draw_barplot(ax,pais,idx)
    ax.figure.canvas.draw_idle()

    ax.figure.canvas.draw_idle()

fig, ax = plt.subplots(figsize=(12,4))

draw_barplot(ax,'','')
#NO QUIERO PINTAR EL NOMBRE DEL PAIS EN EL EJE X
xax = ax.axes.get_xaxis()
xax = xax.set_visible(False)

cid = fig.canvas.mpl_connect('button_press_event', onpick)

plt.title('AQUI MOSTRARÉ LA MEDIA SALARIAL DEL PAIS SELECCIONADO')
plt.show()

```

En el gráfico superior podemos interactuar con él, y consultar el salario medio de cada país. Si bien es cierto que el nivel de interacción y exploración de los datos no es del todo amigable. Es por ello que a partir de aquí aumento la interacción entre el usuario y los datos<br>
Lo siguiente es poder consultar el salario medio mensual a nivel de ciudades, una vez se seleccione el país.

```python

fig.canvas.mpl_disconnect(cid)

listaPaises = list(paises)

def plotCountrySalary(pais):
    df_ = df.loc[df['country'] == pais]
    #ELIMINAR REGISTROS SI "Average Monthly Net Salary (After Tax) (USD)" ES NULL
    df_ = df_.dropna(subset=['Average Monthly Net Salary (After Tax) (USD)'])
    
    fig3, ax3 = plt.subplots(figsize=(12,4))
    
    xax = ax3.axes.get_xaxis()
    xax = xax.set_visible(False)
    
    
    sns.scatterplot(x="city", y="Average Monthly Net Salary (After Tax) (USD)", data=df_)
    
    cid = fig3.canvas.mpl_connect('button_press_event', lambda event: onpickCountry(event, df_, ax3, pais))

    plt.show()
    
    
def onpickCountry(event, df_, ax, pais):
    
    idx = int(event.xdata.round())
    ciudad = df_.iloc[idx, df_.columns.get_loc('city')]
    salario = df_.iloc[idx, df_.columns.get_loc('Average Monthly Net Salary (After Tax) (USD)')]
    
    cols = ['red' if (x == ciudad) else 'blue' for x in df_.city ]
    sns.scatterplot(ax=ax, x="city", y="Average Monthly Net Salary (After Tax) (USD)", data=df_, picker=True, c=cols, pickradius=0.1)
    
    info = 'EN ' + pais + ' EL SALARIO MEDIO MENSUAL EN ' + ciudad + ' ES DE ' + str(salario) + ' DOLARES'
    plt.title(info)
    ax.figure.canvas.draw_idle()
```

```python
import ipywidgets

ipywidgets.interact(plotCountrySalary, pais=listaPaises)
```

Gráfico interactivo donde se puede seleccionar el artículo. A continuación se muestra un plot donde cada punto representa el precio del artículo de cada muestra. <br>
Es posible cruzarlo con el salario medio mensual mediante el cambio de color de los puntos, de manera que se pueda la variación del precio del artículo VS salario medio mensual.<br>
En función del artículo seleccionado, por ejemplo, el precio de un kilo de filetes pollo, se puede ver que el precio es mayor a medida que el salario medio aumenta.

```python
listaProductos = df.columns[3:56]

def costeVida(salario,producto):
    
    fig, ax = plt.subplots(figsize=(12,4))
    
    #PRECIO DEL PRODUCTO SELECCIONADO DONDE EL SALARIO SEA IGUAL O INFERIOR AL SALARIO RECIBIDO POR LA FUNCION
    df_ = df.loc[df['Average Monthly Net Salary (After Tax) (USD)'] >= 0]
    df_ = df_.filter([producto,'Average Monthly Net Salary (After Tax) (USD)'])
    df_ = df_.dropna(subset=[producto])
    df_.rename(columns={producto:'Precio', 'Average Monthly Net Salary (After Tax) (USD)':'Salario'}, inplace=True)
    
    colors = ['green' if (x >= salario) else 'blue' for x in df_.Salario ]
    sns.scatterplot(data=df_['Precio'], c=colors)
    
    
ipywidgets.interact(costeVida, salario=(1,10000,10), producto=listaProductos)
```

Ahora voy a analizar las ciudades más caras/baratas (hasta un máximo de 20). Para ello, podremos seleccionar el artículo, y pintaremos aquellas N ciudades donde dicho artículo es más caro

```python
opciones = ['caro', 'barato']
def ciudadesCaras(producto,numCiudades, orden):
    
    if orden == 'caro':
        X=False
    else:
        X=True
    
    fig, ax = plt.subplots(figsize=(12,4))
    df_ = df.filter(['city',producto])
    df_ = df_.sort_values(by=[producto], ascending=X).iloc[0:numCiudades]
    
    sns.barplot(y='city',x=producto,data=df_)

ipywidgets.interact(ciudadesCaras, producto=listaProductos, numCiudades=(1,20,1), orden=opciones)
```

Procedo a analizar visualmente el impacto que representa el precio de una serie de productos sobre el salario medio mensual. Es posible seleccionar la ciudad, producto y cantidad.

```python
city = df.city.unique()
city.sort()

productos = [
                 'Water (0.33 liter bottle, in restaurants) (USD)',
                 'Milk (regular), (1 liter) (USD)',
                 'Loaf of Fresh White Bread (500g) (USD)',
                 'Rice (white), (1kg) (USD)',
                 'Eggs (regular) (12) (USD)',
                 'Chicken Fillets (1kg) (USD)',
                 'Beef Round (1kg) (or Equivalent Back Leg Red Meat) (USD)',
                 'Apples (1kg) (USD)',
                 'Tomato (1kg) (USD)',
                 'Potato (1kg) (USD)',
                 'Gasoline (1 liter) (USD)',
                 'Basic (Electricity, Heating, Cooling, Water, Garbage) for 85m2 Apartment (USD)']

def analisis(ciudad, cantidad, producto = productos):
    
    # DEL DF PRINCIPAL SELECCIONAR EL PAIS,PRODUCTO Y SALARIO MENSUAL
    df_ = df.loc[df['city'] == ciudad]

    df_ = df_.filter(['city',producto,'Average Monthly Net Salary (After Tax) (USD)'])
    
    salario = df_['Average Monthly Net Salary (After Tax) (USD)'].values[0]
    precio = df_[producto].values[0] * cantidad
    info = [precio,salario]
    
    fig, ax = plt.subplots(figsize=(12,4))
    
    plt.pie(info, labels=[producto, 'Average Monthly Net Salary (After Tax) (USD)'], autopct='%.0f%%')
    
ipywidgets.interact(analisis, ciudad=city, cantidad=(1,200,1), producto = productos)
```
